```python
#Categorical and numerical variable barcha
region = df.groupby('Region')['Sales'].sum().round(2).index 
tot_sales= df.groupby('Region')['Sales'].sum().round(2). values

plt.figure(figsize=[15,20])
base_color= sns.color_palette()[0]

plt.subplot(3,2,1)
ax= sns.barplot(data=df,x='Region',y='Sales', color= base_color,ci=None)
plt.title('Barplot of mean sales per region')
plt.xlabel('Region')
plt.ylabel('Mean Sales')
ax.bar_label(ax.containers[0])

plt.subplot(3,2,2)
ax1= sns.countplot(data=df,x='Region', color=base_color)
ax1.bar_label(ax1.containers[0])
plt.title('Barplot of counts per region')
plt.xlabel('Region')
plt.ylabel('Counts')

plt.subplot(3,2,3)
ax=sns.barplot(data=df, x='Region', y='Sales', color= base_color, estimator=np.sum, ci=None)
plt.title('Barplot of total sales per region')
ax.bar_label(ax.containers[0])
ax.set_xlabel('Region')
ax.set_ylabel('Total Sales')

plt.subplot(3,2,4)
ax2= sns.histplot(data=df, x= 'Quantity', color= 'g', bins= 10)
ax2.bar_label(ax2.containers[0])
ax2.set_title('Histogram for Quantity')
ax2.set_xlabel('Quantity')
ax2.set_ylabel('Counts')

plt.subplot(3,2,5)
ax2= sns.barplot(data= df, x='Region', y='Sales', color= base_color, estimator=np.median)
ax2.bar_label(ax2.containers[0])
ax2.set_title('Barplot of Sales Against Region')
ax2.set_xlabel('Region')
ax2.set_ylabel('Median Sales')

plt.subplot(3,2,6)
ax2= sns.barplot(data=df, x='Region', y= 'Sales', hue= 'Category', estimator=np.sum, ci= None)
ax2.set_title('Multivariate Barplot')
ax2.set_xlabel('Region')
ax2.set_ylabel('Total Sales')

plt.subplots_adjust(wspace=0.2, hspace=0.2, bottom=0.1);



               
               
               
         
```
