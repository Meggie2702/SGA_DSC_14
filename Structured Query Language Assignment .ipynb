1)create a table named "employees" with columns such as employee_id(int): make it the primary key, last_name, first_name, pisition, salary, date_hired.

```python
import sqlite3 

conn= sqlite3.connect("employees.db")

cursor= conn.cursor()

print("cursor created successfully \n", type (cursor))
```

```python
cursor.execute(""" CREATE TABLE employees(
employee_id INT AUTO_INCREMENT,
last_name TEXT,
first_name TEXT,
position TEXT ,
salary INT,
date_hired DATE,
PRIMARY KEY(employee_id)
)""")
print("TABLE SUCCESSFULLY CREATED")

conn.commit()

```
