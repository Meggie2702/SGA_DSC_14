**QUESTION 2:
Using matplotlib module, make a lineplot of 'age' on the x-axis against 'children' on the y-axis using 
the following labels for your plot:
x-axis: 'Age (years)'
y-axis: 'Number of children'
plot title: 'Plot of Age against children'.**

```python
import csv
x=[]
y=[]
with open('/storage/emulated/0/Download/insurance.csv', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(row[0])
        y.append(row[3])
                 
plt.plot(x,y, label= "Age", color= 'g', linestyle= 'dashed', marker= 'o')

plt.xlabel('Age(years)')
plt.ylabel('Number of children')
plt.title('Plot of Age against Children', fontsize =16)
plt.grid
plt.legend();
     
        
        
```

**QUESTION 3:
How would you use matplotlib module to make two different subplots on different axes with the following 
characteristics?**

**a) First subplot: 
i. pie chart using the 'sex' column. Ensure the percentage on the pie chart is taken to 2 d.p.

**b) Second subplot:
i. Create a bar-chart with the 'region' column. 
ii. Carefully label the chart with x-axis: 'Region'; y-axis: 'Counts'; title: 'Bar plot for Region'.

```python
#a) First subplot: i. pie chart using the 'sex' column. Ensure the percentage on the pie chart is taken to 2 d.p
x= df.sex.value_counts()

sex = df.sex.unique()

plt.pie(x, labels = sex, autopct='%.2f%%')
plt.title('Sex', fontsize=16)


```

#b) Second subplot: i. Create a bar-chart with the 'region' column. ii. Carefully label the chart with x-axis: 'Region'; y-axis: 'Counts'; title: 'Bar plot for Region'.

```python
plt.title('Bar Plot For Region', fontsize=16)
plt.figure(figsize=[7,4])
plt.bar(df.region, color= 'brown', height= 14)
plt.xlabel('Region')
plt.ylabel('Counts')

```

**QUESTION 4:Using matplotlib, plot two histograms on the same axis for columns 'age' and 'children'. Ensure the following
is indicated in the plot:**
1. The two plots should be clearly visible to the eye [Hint: Use the 'alpha' argument]
ii. Use appropriate titles for the axes and the plot

**Using the dataset given below,**

**x_var = np.random.normal(loc = 40, scale = 1.5, size = 200)

z_var = np.random.normal(loc = 60, scale = 2, size = 300)**

**i. Create histograms for the two variables,x_var and y_var, on the same axis. Ensure the two plots can easily 
be see using the appropriate parameter. Also, use the legend() function for the plot.**

```python
x_var = np.random.normal(loc = 40, scale = 1.5, size = 200)
y_var = np.random.normal(loc = 50, scale = 1.5, size = 300)
z_var = np.random.normal(loc = 60, scale = 2, size = 300)

plt.figure(figsize=[7,4])
plt.title(' My X and Y Histogram', fontsize= 16)
plt.hist(x_var, bins= 10, alpha=0.5, label='X axis')
plt.hist(y_var, bins=10, alpha= 0.6, label= 'Y axis')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend();

```

**ii. Make two subplots using the 3 variables as follows:**

1. Make a lineplot of y_var against z_var for the first subplot
2. Make a histogram for the second subplot using x_var.

```python
x_var = np.random.normal(loc = 40, scale = 1.5, size = 200)
y_var = np.random.normal(loc = 50, scale = 1.5, size = 300)
z_var = np.random.normal(loc = 60, scale = 2, size = 300)

fig,(ax1,ax2) = plt.subplots(nrows=1,ncols=2,figsize=[7,4])

#1st Subplot
ax1.set_title('Subplot 1')
ax1.plot(y_var, z_var, color='green')
ax1.set_xlabel('X axis')
ax1.set_ylabel('Z axis')

#2nd Subplot
ax2.set_title('Subplot 2')
ax2.hist( x_var,color= 'blue')
ax2.set_xlabel('X axis')


                        
```
